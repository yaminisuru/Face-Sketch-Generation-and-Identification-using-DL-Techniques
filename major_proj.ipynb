```python
import os

SKETCH_DIR =  "C:\\Users\\durga\\OneDrive\\Desktop\\face_sketch_project\\dataset\\sketches"

for file in os.listdir(SKETCH_DIR):
    if "-sz1" in file:
        new_name = file.replace("-sz1", "")
        os.rename(
            os.path.join(SKETCH_DIR, file),
            os.path.join(SKETCH_DIR, new_name)
        )

print("Sketch files renamed successfully")
```

```python
import os

SKETCH_DIR = "C:\\Users\\durga\\OneDrive\\Desktop\\face_sketch_project\\dataset\\sketches"

for file in os.listdir(SKETCH_DIR):
    if file.startswith("M2"):
        new_name = file.replace("M2", "m", 1)

        os.rename(
            os.path.join(SKETCH_DIR, file),
            os.path.join(SKETCH_DIR, new_name)
        )

print("✅ All M2 filenames changed to m successfully!")
```
